## Quantum Machine Learning (QSVM)
---

* This project builds a Quantum Support Vector Machine (QSVM), a hybrid algorithm that combines classical machine learning with quantum computing.

* The core idea is to replace the classical "kernel"—the part of an SVM that maps data to higher dimensions—with a quantum circuit. We will encode classical data (e.g., from a simple "moons" dataset) into quantum states, and then use a quantum computer simulator (via Qiskit) to measure the "distance" between these states in a high-dimensional quantum space.

* The goal is to see if this "quantum kernel" can find patterns and separate data points that a purely classical SVM finds difficult, demonstrating a potential quantum advantage in machine learning.
